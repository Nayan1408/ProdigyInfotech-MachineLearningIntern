```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Population of India.csv',  encoding='utf-8')
df.head()
```

```python

plt.figure(figsize=(12, 8))

plt.bar(df['State/UT'], df['Female'], color='skyblue')
plt.xlabel('State')
plt.ylabel('Female Population')
plt.title('Female Population Across States')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```
